### 构建模型

```python
import keras.layers as KL
import keras.utils as KU
import keras.backend as K
from keras.models import Model
```

```python
os.environ['CUDA_VISIBLE_DEVICES'] = '1'
```

```python
start_units = 64

inputs_sen1 = KL.Input([32, 32, 8], name='in_sen1')
inputs_sen2 = KL.Input([32, 32, 10], name='in_sen2')

x1 = KL.Conv2D(start_units, (3, 3), padding='SAME')(inputs_sen1)
# x1 = KL.BatchNormalization()(x1)
x1 = KL.Activation('relu')(x1)
x1 = KL.Conv2D(start_units, (3, 3), padding='SAME')(x1)
# x1 = KL.BatchNormalization()(x1)
x1 = KL.Activation('relu')(x1)
x1 = KL.Dropout(0.1)(x1)
x1 = KL.MaxPooling2D(padding='SAME')(x1)

x2 = KL.Conv2D(start_units, (3, 3), padding='SAME')(inputs_sen2)
# x2 = KL.BatchNormalization()(x2)
x2 = KL.Activation('relu')(x2)
x2 = KL.Conv2D(start_units, (3, 3), padding='SAME')(x2)
# x2 = KL.BatchNormalization()(x2)
x2 = KL.Activation('relu')(x2)
x2 = KL.Dropout(0.1)(x2)
x2 = KL.MaxPooling2D(padding='SAME')(x2)

x = KL.Concatenate()([x1, x2])
x = KL.Conv2D(start_units * 2, (3, 3), padding='SAME')(x)
# x = KL.BatchNormalization()(x)
x = KL.Activation('relu')(x)
x = KL.Dropout(0.1)(x)
x = KL.Conv2D(start_units * 2, (3, 3), padding='SAME')(x)
# x = KL.BatchNormalization()(x)
x = KL.Activation('relu')(x)
x = KL.Dropout(0.1)(x)
x = KL.MaxPooling2D(padding='SAME')(x)

x = KL.Conv2D(start_units * 4, (3, 3), padding='SAME')(x)
# x = KL.BatchNormalization()(x)
x = KL.Activation('relu')(x)
x = KL.Dropout(0.1)(x)
x = KL.Conv2D(start_units * 4, (3, 3), padding='SAME')(x)
# x = KL.BatchNormalization()(x)
x = KL.Activation('relu')(x)
x = KL.Dropout(0.1)(x)
x = KL.Conv2D(start_units * 4, (3, 3), padding='SAME')(x)
# x = KL.BatchNormalization()(x)
x = KL.Activation('relu')(x)
x = KL.MaxPooling2D(padding='SAME')(x)

x = KL.Flatten()(x)
x = KL.Dropout(0.1)(x)
x = KL.Dense(start_units * 8, activation='relu')(x)
output = KL.Dense(17, activation='softmax', name='prediction')(x)
```

```python
train_model = Model(inputs=[inputs_sen1, inputs_sen2], outputs=output)
```

```python
train_model.summary()
```

```python
from keras.optimizers import Adam
train_model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=1e-4, clipnorm=0.001), metrics=[my_acc_metric_tf])
```

```python
train_model.fit_generator(data_train,
                          epochs=10,
                          steps_per_epoch=20000,
                          validation_data=data_val,
                          validation_steps=2000)
```

```python
train_model.save_weights('./snapshots/tsnet-9/tsnet_best.h5')
```

### 验证集精度

```python
def my_acc_metric_tf(y_true, y_pred):
    '''
    y_true : [None, 17]
    y_pred : [None, 17]
    '''
    y_true = K.argmax(y_true, axis=-1)
    y_pred = K.argmax(y_pred, axis=-1)
    acc = K.equal(y_true, y_pred)    # [None, 1]
    acc = K.cast(acc, K.floatx())
    acc = K.mean(acc)
    
    return acc


def my_acc_metric(y_true, y_pred):
    '''
    y_true : [None, 17]
    y_pred : [None, 17]
    '''
    y_true = np.argmax(y_true, axis=-1)
    y_pred = np.argmax(y_pred, axis=-1)
    acc = np.equal(y_true, y_pred)    # [None, 1]
    acc = acc.astype(np.float32)
    acc = np.mean(acc)
    
    return acc
```

```python
m = []
data_val.index = 0
for i in range(data_val.count):
    if i % 10 == 0:
        print('processing : {0}/{1}'.format(i, data_val.count), end='\r')
    [x1, x2], y = next(data_val)
    ret = train_model.predict([x1, x2])
    m.append(my_acc_metric(y, ret))
print('\n', np.mean(m))
```

```python
class generator():
    def __init__(self,
                 filepath='/home/lmzwhu/programs/DATAS/LCZ/training.h5',
                 batch_size=8,
                 datatype='train',
                 split=0.1):
        data_file = h5py.File(filepath, 'r')

        self.X1 = data_file['sen1']
        self.X2 = data_file['sen2']
        self.Y = data_file['label']

        self.indices = np.arange(self.Y.shape[0] // batch_size)
        np.random.seed(1)
        np.random.shuffle(self.indices)
        split = round(self.Y.shape[0] // batch_size * split)
        if datatype == 'train':
            self.indices = self.indices[:-split]
        else:
            self.indices = self.indices[-split:]

        self.count = self.indices.size
        self.batch_size = batch_size
        self.index = 0

    def next_batch(self, batch_size=8):
        idx = self.indices[self.index]
        images_1 = self.X1[idx:idx+batch_size]
        images_2 = self.X2[idx:idx+batch_size]
        labels = self.Y[idx:idx+batch_size]

        self.index += batch_size
        if self.index >= self.count:
            self.index = 0
            np.shuffle(self.indices)

        images_1 = np.asarray(images_1, dtype=np.float32)
        images_2 = np.asarray(images_2, dtype=np.float32)
        labels = np.asarray(labels, dtype=np.float32)

        return [images_1, images_2], labels

    def __next__(self):
        return self.next_batch(self.batch_size)
```
